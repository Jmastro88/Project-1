```python
#dependencies

import pandas as pd
```

```python
#import datasources from previously created CSV - GDP, Average MPG and Annual Miles

file1="./Resources/GDP_AvgMPG_TotalMiles.csv"
gdp_mpg_miles=pd.read_csv(file1,encoding = "ISO-8859-1",header=0)
```

```python
#import datasources from previously created pricing/consumption csv file
file2="./Resources/price_cons.csv"
price_cons=pd.read_csv(file2,encoding = "ISO-8859-1",header=0)
price_cons.rename(columns = {'MSN':'Factors'}, inplace = True)
price_cons
```

```python
#concatenate dataframes to form consolidated dataframe
con_data=pd.concat([gdp_mpg_miles,price_cons])
#set index to 
con_data = con_data.set_index(['Factors'])
con_data
```

```python
con_data.to_csv('Summary File for Analysis.csv')
```
