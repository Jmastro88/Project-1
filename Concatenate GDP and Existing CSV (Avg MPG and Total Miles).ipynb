```python
#dependencies

import pandas as pd
```

```python
#import datasources from created CSV - Average MPG and Annual Miles

file1="./Resources/average mpg and total miles"
avg=pd.read_csv(file1,encoding = "ISO-8859-1",header=0)
```

```python
#import datasources from FRED GDP csv file
file2="./Resources/GDPCA.csv"
gdp=pd.read_csv(file2,encoding = "ISO-8859-1",header=0)

#restrict to relevant time series (2007 to 2017)
gdp_rest = gdp[['Unnamed: 0','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017']]
gdp_rest
```

```python
#concatenate dataframes to form consolidated dataframe
con_data=pd.concat([gdp_rest,avg])
con_data.rename(columns = {'Unnamed: 0':'Factors'}, inplace = True)
#set index to 
con_data = con_data.set_index(['Factors'])
con_data
```

```python
con_data.to_csv('GDP_AvgMPG_TotalMiles.csv')
```
