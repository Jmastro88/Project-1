```python
#dependencies

import pandas as pd
```

```python
#import datasources from CSV- Average MPG and Annual Miles

file="./Resources/Avg MPG.csv"

avg_mpg=pd.read_csv(file,encoding = "ISO-8859-1",header=1)

avg_mpg
```

```python
#import datasources from CSV - Average MPG and Annual Miles
file = "./Resources/Annual Miles.csv"
annualmiles = pd.read_csv(file,encoding = "ISO-8859-1",header=1)
annualmiles
```

```python
annualmiles.set_index('Unnamed: 0')
annualmiles_rest = annualmiles.loc[3:3]
annualmiles_rest


# merged_data=pd.DateFrame({})

# avg_light
```

```python
avg_light=avg_mpg[avg_mpg["Unnamed: 0"]== "Average U.S. light duty vehicle fuel efficiency (mpg) (calendar year)"]

avg_light
```

```python
con_data=pd.concat([annualmiles_rest,avg_light])
cols = list(con_data.columns)
cols = [cols[-1]] + cols[:-1]
con_data = con_data[cols]
con_data = con_data.set_index(['Unnamed: 0'])
con_data
```

```python
con_data_rest = con_data[['2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017']]
con_data_rest
con_data_rest.to_csv('average mpg and total miles')
```
